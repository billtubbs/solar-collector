```python
import casadi as ca

# Create an optimization problem
opti = ca.Opti()

# Define variables
x = opti.variable()
y = opti.variable()

# Define objective
opti.minimize(x + y)

# Define constraints
opti.subject_to(x + 2 * y >= 4)
opti.subject_to(x - y <= 1)
opti.subject_to(x >= 0)
opti.subject_to(y >= 0)

# Select solver
opti.solver("ipopt", {"print_time": False}, {"print_level": 0})

# Solve the problem
sol = opti.solve()

# Display results
print("Status:", sol.stats()["return_status"])
print("Optimal x:", sol.value(x))
print("Optimal y:", sol.value(y))
print("Optimal Objective:", sol.value(x + y))
```
